```python
yourTag = "hiking"
filename = 'instacrawl_' + yourTag + '.json'
```

```python
# items = list(set(items))
print(len(items))
data = []
for item in items:
  try: 
      text = item.text.split("\n")[0]
      words = text.split()
      not_tags = []
      tags = []
      for word in words:
        if '#' in word:
          #태그 # 삭제
          tags.append(word.strip("#"))
        else:
          not_tags.append(word)
      caption = ' '.join(e for e in not_tags)
      image = item.find_element(by=By.CLASS_NAME, value = "post-image")
      x = {
          "label": yourTag,
          "img_name": image.get_attribute('src'),
          "text": caption,
          "hashtag": tags
      }      
      # print(x)
      data.append(x)
      
  except:
      pass
print(data)

#json 파일로 저장

with open(filename,'w') as f:
  json.dump(data, f , ensure_ascii=False)
```

### 논문 전처리 과정: 
In the case of texts and hashtags, they were refined for 10,000 instances acquired through the images preprocess. We removed emojis, except for some special symbols, andcharacters such as Chinese or Japanese were removed or translated. In addition, like the images, information such as names, user ids, and phone numbers, that can identify an individual was also removed or modified.
Also, comments and post URLs were removed from the instances.
1. 중복데이터 제거
2. text에 특수기호 제거: 이모지, 숫자, 이름 제거, 외국어

```python
# json 파일 text 파트 전처리: 이모지, 특수기호(온점, 쉼표 이런거는 남기고), 숫자, 영어 모두 제거

import re
for i in range(0,len(temp["text"])):
  temp_str=temp["text"][i]

  emoji_pattern = re.compile("["
          u"\U0001F600-\U0001F64F"  # emoticons
          u"\U0001F300-\U0001F5FF"  # symbols & pictographs
          u"\U0001F680-\U0001F6FF"  # transport & map symbols
          u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                            "]+", flags=re.UNICODE)

  temp_str=emoji_pattern.sub(r'', temp_str)

  temp_str=re.sub(r'[^ㄱ-힗,.!~]','',temp_str)
  temp["text"][i]=temp_str
```
